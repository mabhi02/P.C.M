```python
from PIL import Image
import matplotlib.pyplot as plt
import os

def upsample_and_display_image(input_path):
    """
    Upsamples a 224x224 image to 448x448 and displays both versions side by side.
    
    Args:
        input_path (str): Path to the input image file
    """
    try:
        # Open the image
        img = Image.open(input_path)
        
        # Get original dimensions
        original_width, original_height = img.size
        print(f"Original image dimensions: {original_width}x{original_height}")
        
        # Resize to 224x224 if not already that size
        if original_width != 224 or original_height != 224:
            print(f"Resizing from {original_width}x{original_height} to 224x224...")
            img = img.resize((224, 224), Image.LANCZOS)
        
        # Upsample to 448x448
        upsampled_img = img.resize((448, 448), Image.LANCZOS)
        
        # Create output path for the upsampled image
        file_dir, file_name = os.path.split(input_path)
        name, ext = os.path.splitext(file_name)
        output_path = os.path.join(file_dir, f"{name}_upsampled{ext}")
        
        # Save the upsampled image
        upsampled_img.save(output_path)
        print(f"Upsampled image saved as: {output_path}")
        
        # Display both images side by side
        plt.figure(figsize=(12, 6))
        
        plt.subplot(1, 2, 1)
        plt.imshow(img)
        plt.title("Original (224x224)")
        plt.axis('off')
        
        plt.subplot(1, 2, 2)
        plt.imshow(upsampled_img)
        plt.title("Upsampled (448x448)")
        plt.axis('off')
        
        plt.tight_layout()
        plt.show()
        
        print("Images displayed successfully")
        
    except Exception as e:
        print(f"Error: {str(e)}")

if __name__ == "__main__":
    # Hardcoded file path for the specific image
    input_image_path = r"C:\Users\athar\Documents\GitHub\P.C.M\ai_vs_human_gener\train_data\train_image_1_Human.png"
    
    # Call the function with the hardcoded path
    upsample_and_display_image(input_image_path)
```
